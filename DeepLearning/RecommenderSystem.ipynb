The dataset consists of 943 users and 1682 movies in which each user has rated at least 20 movies. The training and testing data are pre-processed as follows: we apply the database to build a U by I matrix A, where U is the number of users, and I is the number of rated movies. Each element A_ui denotes the rating scored by the u-th user for the i-th movie.

```python
def buildMatrixA(filename, rows, cols):
    A = [[0]*cols for i in range(rows)]
    with open(filename,'r') as datafile:
        for line in datafile:
            instance = line.split('\t')
            u = int(instance[0]) - 1
            i = int(instance[1]) - 1
            rating = int(instance[2])
            A[u][i] = rating
    return A
```

```python
A = buildMatrixA('u.data',943,1682)
```

Build a recommender system where the baseline predictor for a (u,i) pair is: b_ui=μ+b_u+b_i as described in [2] section 4.1.

a) First, we will calculate μ, the average movie rating in the training set.

```python
A_uabase = buildMatrixA('ua.base',943,1682)
```

```python
def getAvgRating(matrix):
    total = 0
    count = 0
    for u in matrix:
        for i in range(len(matrix[0])):
            if u[i] > 0:
                total += u[i]
                count += 1
    return float(total)/count
```

```python
mu1 = getAvgRating(A_uabase)
print 'Average rating for ua.base: %f' % mu1
```

Now, we can use decoupling to create the vector of bi: 
bi = (∑u∈R(i)(rui − µ))/(λ2 +|R(i)|)

```python
def make_bi(matrix, mu):
    # first, get vector of |R(i)| values
    ri = [0]*len(matrix[0])
    for u in matrix:
        for i in range(len(u)):
            if u[i] > 0:
                ri[i] += 1
    bi = [0]*len(ri)
    # now, compute bi
    for u in matrix:
        for i in range(len(u)):
            if u[i] > 0:
                bi[i] += (u[i] - mu)
    L2 = 25
    for x in range(len(bi)):
        bi[x] /= (L2 + ri[x])
    return bi
```

```python
bi1 = make_bi(A_uabase, mu1)
```

Then, for each user u we set bu = ∑i∈R(u) (rui − µ −bi)/ (λ3 +|R(u)|)
.

```python
def make_bu(matrix, bi, mu):
    # first, get vector of |R(u)| values
    ru = [0]*len(matrix)
    for u in range(len(matrix)):
        for i in range(len(matrix[0])):
            if matrix[u][i] > 0:
                ru[u] += 1
    # now, compute bu
    bu = [0]*len(matrix)
    for u in range(len(matrix)):
        for i in range(len(matrix[0])):
            if matrix[u][i] > 0:
                bu[u] += (matrix[u][i] - mu - bi[i])
    L3 = 10
    bu = [bu[x]/(L3 + ru[x]) for x in range(len(bu))]
    return bu
```

```python
bu1 = make_bu(A_uabase, bi1, mu1)
```

Now we have all the elements needed for the baseline predictor bui=μ+bu+bi. We can now compute the means square error on the test set. First, we will create simple helper function that implements the predictor.

```python
def predictor(mu, u, i, bu, bi):
    return mu + bu[u] + bi[i]
```

Now we can create the matrix for the test set data, and get the RMSE.

```python
A_uatest = buildMatrixA('ua.test',943,1682)
```

```python
# compute RMSE for the test set
def RMSE(matrix, mu, bu, bi):
    rmse = 0
    size = 0
    for u in range(len(matrix)):
        for i in range(len(matrix[0])):
            if matrix[u][i] > 0:
                rmse += (matrix[u][i] - predictor(mu,u,i,bu,bi))**2
                size += 1
    rmse = (float(rmse)/size)**0.5
    return rmse
```

```python
rmse1 = RMSE(A_uatest, mu1, bu1, bi1)
print 'The RMSE for ua.test is: %f' %rmse1
```

We can repeat this process for ub.base and ub.test.

```python
A_ubbase = buildMatrixA('ub.base',943,1682)
mu2 = getAvgRating(A_ubbase)
bi2 = make_bi(A_ubbase, mu2)
bu2 = make_bu(A_ubbase, bi2, mu2)
A_ubtest = buildMatrixA('ub.test',943,1682)
rmse2 = RMSE(A_ubtest, mu2, bu2, bi2)
print 'The RMSE for ub.test is: %f' %rmse2
```

The RMSE for each movie rating can also be calculated. To do so, a new function to filter the matrix based on rating is created.

```python
# filter matrix to only contain given rating
def filter_rating(matrix, rating):
    filtered = [[0]*len(matrix[0]) for x in range(len(matrix))]
    for u in range(len(matrix)):
        for i in range(len(matrix[0])):
            if matrix[u][i] == rating:
                filtered[u][i] = rating
            else:
                filtered[u][i] = 0
    return filtered
```

We will now get the average rating for every user. Using this, we will group users into 5 buckets based on their average rating and compute RSME for each group.

```python
# get average rating for every user
def avgUserRating(matrix):
    avgs = [0]*len(matrix) #one cell per user
    for u in range(len(matrix)):
        count = 0
        for i in range(len(matrix[0])):
            if matrix[u][i] > 0:
                avgs[u] += matrix[u][i]
                count += 1
        avgs[u] = float(avgs[u])/count
    return avgs
```

```python
user_avgs = avgUserRating(A_uatest)
print 'Min average rating for a user is: %.2f' % min(user_avgs)
print 'Max average rating for a user is: %.2f' % max(user_avgs)
```

```python
# filter matrix to include only ratings of users with average in specified range
def filter_users(matrix, userAvgs, low, high):
    filtered = [[0]*len(matrix[0]) for x in range(len(matrix))]
    for u in range(len(matrix)):
        if (userAvgs[u] >= low) and (userAvgs[u] < high):
            for i in range(len(matrix[0])):
                filtered[u][i] = matrix[u][i]
    return filtered
```

Build a recommender system based on [2] section 4.2. For a (u,i) pair compute the rating by using the item similarity function.
First, a helper function will be made to get the set of users that has rated both items i1 and i2. Then we can construct the similarity function.

```python
def UsersRatingBothItems(matrix, i1, i2):
    users = []
    for u in range(len(matrix)):
        if matrix[u][i1] > 0 and matrix[u][i2] > 0:
            users.append(u)
    return users
```

```python
def similarity(matrix, userAvgs, i1, i2):
    users = UsersRatingBothItems(matrix, i1, i2)
    sum1, sum2, sum3 = 0,0,0
    for user in users:
        #print 'user: %i i1: %i i2: %i' %(user,i1,i2)
        diff1 = matrix[user][i1] - userAvgs[user]
        diff2 = matrix[user][i2] - userAvgs[user]
        sum1 += (diff1 * diff2)
        sum2 += diff1**2
        sum3 += diff2**2
    denominator = (sum2 * sum3)**0.5
    if denominator > 0:
        sim = float(sum1)/denominator
    else:
        sim = 0
    return sim
```

```python
def similarityMatrix(matrix, userAvgs):
    size = len(matrix[0])
    SM = [[0]*size for x in range(size)]
    for i in range(size):
        for i2 in range(size):
            SM[i][i2] = similarity(matrix, userAvgs, i, i2)
    return SM
```

Now we can construct the pairwise similarity matrix.

```python
user_avgs_base = avgUserRating(A_uabase)
similarity_Matrix = similarityMatrix(A_uabase, user_avgs_base)
```

For a (u,i) pair find the k most similar items to i that the user u has rated.

```python
import numpy as np
```

```python
def knn(matrix, u, i, k, simMatrix):
    simVector = [0]*len(matrix[0])
    for item in range(len(matrix[0])):
        if matrix[u][item] > 0:
            simVector[item] = simMatrix[i][item]
    return np.argsort(simVector)[-k:]
```

The weighted sum method computes the prediction on an item i for a user u by computing the sum of the ratings given by the user on the items similar to i. Each ratings is weighted by the corresponding similarity si;j between items i and j.

```python
def weightedSum(matrix, u, i, k, simMatrix, globalAvgRtng):
    numerator = 0
    denominator = 0
    simItems = knn(matrix, u, i, k, simMatrix)
    for item in simItems:
        numerator += (simMatrix[i][item] * matrix[u][item])
        denominator += abs(simMatrix[i][item])
    if denominator > 0:
        result = float(numerator)/denominator
    else:
        result = globalAvgRtng
    return result
```

```python
# compute RMSE for the test set using the weightedSum predictor
def RMSE2(matrix, simMatrix, k):
    rmse = 0
    size = 0
    globalAvgRtng = getAvgRating(matrix)
    for u in range(len(matrix)):
        for i in range(len(matrix[0])):
            if matrix[u][i] > 0:
                rmse += (matrix[u][i] - weightedSum(matrix, u, i, k, simMatrix, globalAvgRtng))**2
                size += 1
    rmse = (float(rmse)/size)**0.5
    return rmse
```
